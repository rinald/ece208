```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def bisection(f, x1, x2, n=5, t=0.01):
    _x = np.linspace(x1, x2, 100)
    _0 = np.linspace(0, 0, 100)
    _y = f(_x)
    plt.plot(_x, _y, 'k')
    plt.plot(_x, _0, 'b--')

    plt.plot(x1, f(x1), 'or')
    plt.plot(x2, f(x2), 'or')

    for i in range(n):
        if f(x1)*f(x2) < 0:
            x = (x1+x2)/2
            print(f'p{i+1} = {x}')
            plt.plot(x, f(x), '.r')
            if i < 5:
                plt.annotate(f'{i+1}', (x, f(x)), textcoords='offset points', xytext=(0, 10), ha='center')

            if abs(f(x)) < 10**(-8) or (x2-x1)/2 < t:
                break

            if f(x1)*f(x) < 0:
                x2 = x
            elif f(x1)*f(x) > 0:
                x1 = x
            else:
                break
        else:
            break

    plt.grid()
    plt.show()
```

```python
bisection(lambda x: x**3-2, 0, 2, n=10)
```

```python
bisection(lambda x: x**2-3, -2, -1)
```
