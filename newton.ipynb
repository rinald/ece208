```python
import matplotlib.pyplot as plt
import numpy as np

from numpy import sin, cos, pi, e, exp, log
```

```python
def d_dx(f, h=10**(-6)):
    def df_dx(x):
        return (f(x+h)-f(x))/h
    return df_dx
```

```python
def newton(f, a, b, p0, n=5, t=10**-4):
    f_ = d_dx(f)
    x = np.linspace(a, b, 100)
    y = f(x)
    y_ = f_(x)
    _0 = np.linspace(0, 0, 100)

    plt.plot(x, y, 'k')
    plt.plot(x, y_, 'c-.')
    plt.plot(x, _0, 'b--')

    for i in range(n):
        p = p0 - f(p0)/f_(p0)

        print(f'p{i+1} = {p}')
        plt.plot(p, f(p), '.r')
        if i < 5:
            plt.annotate(f'{i+1}', (p, f(p)), textcoords='offset points', xytext=(0, 10*(-1)**i), ha='center')

        if abs(p-p0) < t:
            break

        p0 = p
```

```python
# Ex.1
newton(lambda x: x**2-6, 2, 3, 1, n=3)
```

```python
# Ex. 5.a
newton(lambda x: x**3-2*x**2-5, 1, 4, 2)
```

```python
# Ex. 5.c
newton(lambda x: x-cos(x), 0, pi/2, 1)
```

```python
newton(lambda x: x*exp(x)-2, 0, 1, 0.5)
```
